```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = '../images/1.bmp'
```

```python
img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
blurred_img = cv2.GaussianBlur(img, (11, 11), 0)
sub_img = img - blurred_img
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.title('Original')
plt.imshow(img, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('Blurred')
plt.imshow(blurred_img, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Sub')
plt.imshow(sub_img, cmap='gray')
```
